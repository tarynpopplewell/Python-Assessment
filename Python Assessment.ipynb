```python
import pandas as pd

df = pd.DataFrame({'ts': ['9802830','9802832','9802830','9802830','9802830','9802830','9802830','9802830',],
                    'post_id':['1','1','2','2','2','3','3','1'],
                   'user_id':['1','1','1','1','1','2','2','3'],
                   'action':['view','reblog','view','reblog','like','view','like','view']
                   },index = [0,1,2,3,4,5,6,7])
df
```

```python
df.loc[df['action'] == 'view', 'is_view'] = 1
df.loc[df['action'] == 'reblog', 'is_reblog'] = 1
df
```

```python
posts = pd.unique(df[['post_id']].values.ravel())
posts = pd.DataFrame(posts)
posts.rename(columns={0:'post_id'}, inplace=True)
posts
```

```python
views=pd.DataFrame(df.groupby('post_id')['is_view'].sum())
views.reset_index(inplace=True)
views
```

```python
reblogs = pd.DataFrame(df.groupby('post_id')['is_reblog'].sum())
reblogs.reset_index(inplace=True)
reblogs
```

```python
posts = posts.merge(views, on='post_id', how='left')
posts = posts.merge(reblogs, on='post_id', how='left')
posts
```

```python
ratios = df.groupby('post_id', as_index=False).sum().eval('Ratio = is_reblog / is_view')
ratios
```
